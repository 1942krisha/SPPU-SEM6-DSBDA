```python
# 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
sns.histplot(x="SepalLengthCm",data=df, kde=True)
plt.show()
sns.histplot(x="SepalWidthCm",data=df, kde=True)
plt.show()
sns.histplot(x="PetalLengthCm",data=df, kde=True)
plt.show()
sns.histplot(x="PetalWidthCm",data=df, kde=True)
plt.show()
```

```python
# 3. Create a box plot for each feature in the dataset
sns.boxplot(x="SepalLengthCm", data=df)
plt.show()
sns.boxplot(x="SepalWidthCm", data=df)
plt.show()
sns.boxplot(x="PetalLengthCm", data=df)
plt.show()
sns.boxplot(x="PetalWidthCm", data=df)
plt.show()
```

```python
sns.boxplot(x="SepalLengthCm",y="Species",data=df)
plt.show()
sns.boxplot(x="SepalWidthCm",y="Species",data=df)
plt.show()
sns.boxplot(x="PetalLengthCm",y="Species",data=df)
plt.show()
sns.boxplot(x="PetalWidthCm",y="Species",data=df)
plt.show()
```

```python
df1=df.drop(columns=["Id"],axis=1)
plt.figure(figsize=(9, 5))
sns.boxplot(data=df1)
plt.title('Box Plot - Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.violinplot(data=df1)
plt.title('Violin Plot - Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.kdeplot(data=df1)
plt.title('Violin Plot - Iris Dataset')
plt.show()
```
