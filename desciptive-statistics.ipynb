```python
age_ranges = [0, 18, 26, 36, 46, 56, df['age'].max() + 1]
age_labels = ['0-18', '18-25', '26-35', '36-45', '46-55', '56+']
sample = pd.cut(df['age'], bins=age_ranges, labels=age_labels, right=False)
sample
```

```python
grouped_data = df.groupby(sample)
for group_name, group_data in grouped_data:
    print("Age Group: ", group_name)
    print(group_data)
    print()
```

```python
summary_stats = grouped_data['income'].agg(['mean', 'median', 'min', 'max', 'std'])
summary_stats
```
